```python
binary_cols = train[cat_cols_missiing].nunique()[train[cat_cols_missiing].nunique()==2].index
binary_cols
```

```python
cat_cols = train[cat_cols_missiing]
nunique = cat_cols.nunique()
```

```python
low_card_cols = nunique[(nunique>=3) & (nunique <=10)].index
```

```python
low_card_cols
```

```python
high_card_cols = nunique[nunique>10].index
```

```python
high_card_cols
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in binary_cols:
    train[col] = le.fit_transform(train[col])
```

```python
for col in binary_cols:
    test[col] = le.fit_transform(test[col])
```

```python
train = pd.get_dummies(train,columns=low_card_cols.union(high_card_cols))
```

```python
test = pd.get_dummies(test,columns=low_card_cols.union(high_card_cols))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
para_dist = {
    "n_estimators": [100,200,300,400,500],
    "max_depth" :[None,5,10,15,20,25],
    "min_samples_leaf":[2,4,6],
    "min_samples_split":[2,5,10],
    'max_features': ['sqrt', 'log2', None]
}
model = RandomForestRegressor()
```

```python
rscv = RandomizedSearchCV(
    estimator=model,
    param_distributions=para_dist,
    n_iter=30,
    random_state=42,
    n_jobs=-1,
    cv=5,
    verbose=1
)
```

```python
rscv.fit(x,y)
```

```python
rscv.best_params_
```

```python
pred = rscv.predict(test)
```

```python
revLog_pred = np.expm1(pred)
```

```python
result['SalePrice'] = revLog_pred
```
